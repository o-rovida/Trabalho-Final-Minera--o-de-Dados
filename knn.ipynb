```python
from tool_box import *
# bibliotecas para manipulação de dados
import pandas as pd

from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans

#bibliotecas para avaliação de modelos
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
dataset = pd.read_csv('dataset/divorce.csv', sep=";")
```

```python
k_means = KMeans(
    n_clusters=2, 
    #número de clusters
    init='k-means++', 
    #método de inicialização
    n_init=10, 
    #número de vezes que o algoritmo será executado com diferentes sementes
    max_iter=300, 
    #número máximo de iterações
    tol=0.0001, 
    #tolerância
    verbose=0,
    #imprime informações
    random_state=None, 
    #semente
    copy_x=True,
    #copia ou não a matriz de dados
    algorithm='auto'
    #algoritmo
)

k_neightbor = KNeighborsClassifier(
    n_neighbors=3, 
    #número de vizinhos
    weights='uniform', 
    #pesos
    algorithm='auto', 
    #algoritmo
    leaf_size=30, 
    #tamanho da folha
    p=2, 
    #potência da métrica de minkowski, quando p=2 é a distância euclidiana
    metric='minkowski',
    #métrica de distância
    metric_params=None,
    #parâmetros da métrica
    n_jobs=None
    #número de jobs paralelos
)
```

```python
k_means.fit(dataset.drop('class', axis=1))

center_1 = k_means.cluster_centers_[0]
center_2 = k_means.cluster_centers_[1]

#df with 3 colluns (features, center_1, center_2)
df = pd.DataFrame(columns=['features', 'center_1', 'center_2'])
for i in range(len(dataset.columns)-1):
    df = df.append({'features': dataset.columns[i], 'center_1': center_1[i], 'center_2': center_2[i]}, ignore_index=True)

df.to_csv('result/centers.csv', index=False, decimal=',', sep=';')
```

```python
X = dataset.drop(['class'], axis=1)
y = dataset['class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

k_neightbor.fit(X_train, y_train)

y_pred = k_neightbor.predict(X_test)

print('Acurácia: ', accuracy_score(y_test, y_pred))
print('Matriz de confusão: \n', confusion_matrix(y_test, y_pred))
print('Relatório de classificação: \n', classification_report(y_test, y_pred))
```
